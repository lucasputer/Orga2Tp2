```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style("darkgrid")
from scipy import stats
```

```python
simd = pd.read_csv('output/salida.caso.tiempos.simd.dat', sep=' ')
c = pd.read_csv('output/salida.caso.tiempos.c.default.dat', sep=' ')
sns.tsplot(time=simd['x'], data=simd['tiempo'], interpolate=True, color="green").set_ylabel('Tiempo (Nanosegundos)')
sns.tsplot(time=c['x'], data=c['tiempo'], interpolate=True, color="blue").set_xlabel('Dimension')
plt.show()
```

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_style("darkgrid")
from scipy import stats
```

```python
simdop = pd.read_csv('output/tiempos_solver_lin_solve_optimo.out', sep=' ')
simd2 = pd.read_csv('output/tiempos_solver_lin_solve_2pixel.out', sep=' ')
simd1 = pd.read_csv('output/tiempos_solver_lin_solve_1pixel.out', sep=' ')

sns.plot(time=simdop['x'], data=simdop['tiempo'], interpolate=True, color="green").set_ylabel('Tiempo (Nanosegundos)')
sns.plot(time=simd2['x'], data=simd2['tiempo'], interpolate=True, color="blue").set_xlabel('Dimension')
sns.plot(time=simd1['x'], data=simd1['tiempo'], interpolate=True, color="red").set_xlabel('Dimension')
plt.show()
```

```python
simd = pd.read_csv('output/salida.caso.tiempos.simd.o1.dat', sep=' ')
c = pd.read_csv('output/salida.caso.tiempos.c.o1.dat', sep=' ')
sns.tsplot(time=simd['x'], data=simd['tiempo'], interpolate=True, color="green").set_ylabel('Tiempo (Nanosegundos)')
sns.tsplot(time=c['x'], data=c['tiempo'], interpolate=True, color="blue").set_xlabel('Dimension')
plt.show()
```
